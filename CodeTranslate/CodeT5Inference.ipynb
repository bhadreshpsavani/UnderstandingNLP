## Inference with CodeT5

In this notebook we will check the inference result for PL to PL task in `Salesforce/codet5-base` model which is mentioned in the [CodeT5: Identifier-aware Unified Pre-trained Encoder-Decoder Models
for Code Understanding and Generation](https://arxiv.org/pdf/2109.00859.pdf) research paper.

More code related information can be found from the official [saleforce repo](https://github.com/salesforce/CodeT5)

We will use checkpoint that is available on hugginface model hub [codet5-base](https://huggingface.co/Salesforce/codet5-base) since it is easy to use and experiement

```python
from transformers import RobertaTokenizer, T5ForConditionalGeneration

tokenizer = RobertaTokenizer.from_pretrained('Salesforce/codet5-large')
model = T5ForConditionalGeneration.from_pretrained('Salesforce/codet5-large')
```

## Task1: Python to C

```python
text = "Translate Python to C: if x==0: x += 1"
input_ids = tokenizer(text, return_tensors="pt").input_ids

# simply generate a single sequence
generated_ids = model.generate(input_ids, max_length=50)
print(tokenizer.decode(generated_ids[0], skip_special_tokens=True))
```
